Code to clean the Calendar data by removing null and low importance data to reduce size as well as removing the time column and the forecast and previous columns that showed what was expected change and what it was before hand

```python
import pandas as pd

# Load the CSV file
file_path = '/content/Calender_data.csv'
df = pd.read_csv(file_path)

# Display original size
print(f"Original size: {df.shape}")

# Remove rows with any null values
df_cleaned = df.dropna()

# Remove the 'time', 'forecast', and 'previous' columns
df_cleaned = df_cleaned.drop(columns=['time', 'forecast', 'previous'])

# Remove rows with 'low' importance
df_cleaned = df_cleaned[df_cleaned['importance'] != 'low']

# Convert 'date' to datetime and remove rows from the year 2025
# Using dayfirst=True to handle formats like 13/01/2020
df_cleaned['date'] = pd.to_datetime(df_cleaned['date'], dayfirst=True)
df_cleaned = df_cleaned[df_cleaned['date'].dt.year != 2025]

# Display new size
print(f"Reduced size: {df_cleaned.shape}")

# Save the cleaned data to a new file
output_path = '/content/Calender_data_cleaned.csv'
df_cleaned.to_csv(output_path, index=False)

print(f"Cleaned file saved to: {output_path}")
display(df_cleaned.head())
```

Clean the data by making all the dates match on all the data sets so they are compatible as removing all the unnecessary data from the NVidia stock history from 1999-2019

```python
import pandas as pd

# Load the NVidia stock history dataset
# Assuming the file is located at /content/NVidia_stock_history.csv
file_path_nvidia = '/content/NVidia_stock_history.csv'
df_nvidia = pd.read_csv(file_path_nvidia)

# Convert the 'Date' column to datetime objects with UTC to handle mixed timezones
df_nvidia['Date'] = pd.to_datetime(df_nvidia['Date'], utc=True)

# Filter out data before 2020 (removing 1999-2019)
df_nvidia = df_nvidia[df_nvidia['Date'].dt.year >= 2020]

# Format the 'Date' column to MM/DD/YYYY
df_nvidia['Date'] = df_nvidia['Date'].dt.strftime('%m/%d/%Y')

# Display the first few rows to verify the format
display(df_nvidia.head())
```

```python
# Save the formatted NVidia data to a new CSV file
output_path_nvidia = '/content/NVidia_stock_history_formatted.csv'
df_nvidia.to_csv(output_path_nvidia, index=False)

print(f"Formatted NVidia file saved to: {output_path_nvidia}")
```
